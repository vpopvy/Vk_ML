```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("intern_task.csv")
```

```python
df
```

```python
df['rank'].value_counts()
```

```python
variances = df.var()

# Определяем колонки с нулевой дисперсией
zero_variance_cols = variances[variances == 0].index
```

```python
# Удаляем колонки с нулевой дисперсией
df = df.drop(columns=zero_variance_cols)
```

```python
# Рассчитываем матрицу корреляции
correlation_matrix = df.corr().abs()

# Создаем маску для верхнего треугольника матрицы
upper_triangle_mask = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(np.bool_)
)
```

```python
threshold = 0.95
correlated_features = [column for column in upper_triangle_mask.columns if any(upper_triangle_mask[column] > threshold)]
```

```python
df = df.drop(columns=correlated_features)
```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(df,test_size=0.3,random_state=42)
```

```python
train = train.sort_values(by=['query_id']).reset_index(drop=True)
test = test.sort_values(by=['query_id']).reset_index(drop=True)
```

```python
train.drop(columns=["rank","query_id"])
y_train = train["rank"]
query_train = train["query_id"]

X_test = test.drop(columns=["rank","query_id"])
y_test = test["rank"]
query_test = test["query_id"]
```

```python
from sklearn.metrics import ndcg_score
```

```python
import catboost as cat

train_pool = cat.Pool(
    data=X_train,
    label=y_train,
    group_id=query_train
)

test_pool = cat.Pool(
    data=X_test,
    label=y_test,
    group_id=query_test
)
```

```python

ranker_ct = cat.CatBoostRanker(
    max_depth =  8,
    eval_metric="NDCG",
    random_seed = 42,
    iterations=300
)
ranker_ct.fit(
    train_pool,
    eval_set = test_pool
)
```

```python
predicted_ranks = ranker_ct.predict(X_test)
ndcg_5_Cat = ndcg_score([y_test],[predicted_ranks])
```

```python
from xgboost import XGBRanker
```

```python
X_train = train.drop(columns=["rank","query_id"])
y_train = train["rank"]
query_train = train["query_id"]

X_test = test.drop(columns=["rank","query_id"])
y_test = test["rank"]
query_test = test["query_id"]
```

```python
XGB_Ranker = XGBRanker(
    max_depth =  8,
    objective = 'rank:ndcg',
    eval_metric=["ndcg"],
    seed = 42
)
XGB_Ranker.fit(
    X_train,
    y_train,
    qid=query_train,
    verbose=True
)
```

```python
predicted_ranks = XGB_Ranker.predict(X_test)
ndcg_5_XGB = ndcg_score([y_test],[predicted_ranks])
```
